```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import Imputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression as LR
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt
import scipy as sp
from sklearn.metrics import accuracy_score
```

```python
# data=pd.read_excel("World Happiness Data.xlsx")
delhiData=pd.read_csv("Happiness index - Form Responses 1.csv")
data=delhiData.drop(['Timestamp'],axis=1).copy()
```

```python
data.head(1)
```

```python
feature='Happiness on 1-10'
col=delhiData[feature].values
for i in range(len(data)):
    if(col[i]>=8):
        data[feature][i]='Very Happy With Life'
    elif(col[i]<=7 and col[i]>=5 ): 
        data[feature][i]='Moderately Happy or OK With Life'
    elif(col[i]>=2 and col[i]<=4):
        data[feature][i]='Unhappy or Unpleased With Life'
    elif(col[i]>=0 and col[i]<=1 ):
        data[feature][i]='Extremely dissatisfied With Life'
```

```python
feature='Friends to count on'
col=delhiData[feature].values
for i in range(len(data)):
    if(col[i]=='Yes'):
        data[feature][i]='Have Friends to Count On'
    else:
        data[feature][i]='No Friends to Count On'
    
```

```python
feature='Freedom of Choice'
col=delhiData[feature].values
for i in range(len(data)):
    if(col[i]=='Satisfied'):
        data[feature][i]='Have Freedom to choose'
    else:
        data[feature][i]='Don\'t have freedom to choose'

    
```

```python
feature='Donated in Past'
col=delhiData[feature].values
for i in range(len(data)):
    if(col[i]=='Yes'):
        # major donation, not smaller one's
        data[feature][i]='Have donated in past'
    else:
        data[feature][i]='Did not donate in past'

    
```

```python
feature='Corruption in Gov'
col=delhiData[feature].values
for i in range(len(data)):
    if(col[i]=='Yes'):
        data[feature][i]='thinks government is very corrupt'
    else:
        data[feature][i]='don\'t think government is corrupt'
    
```

```python
feature='Corruption In Business'
col=delhiData[feature].values
for i in range(len(data)):
    if(col[i]=='Yes'):
        # they think that giants like ONGC, Coal /steel industries, pharma or food or industrial giants pay to influence
        data[feature][i]='thinks large businesses are corrupt'
    else:
        data[feature][i]='don\'t think large businesses are corrupt'
    
    
```

```python
feature='Can people be trusted'
col=delhiData[feature].values
for i in range(len(data)):
    if(col[i]=='People have to be dealt wirh carefully'):
        data[feature][i]='Don\'t trust people easily'
    else:
        data[feature][i]='Have trust and faith in people'
    
            
```
